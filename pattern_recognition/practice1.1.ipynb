## Задача 1.1. Меры расстояния и их применимость

**Условие.**
Перечислите и кратко охарактеризуйте следующие меры расстояния между объектами:

1. Евклидово расстояние. Может быть представлено как длина отрезка между точками x и y $$d_{Eu}(x,y)=\sqrt{\sum_{i=1}^N (x_i-y_i)^2}$$
2. Манхэттенское расстояние. Сумма растояний по проекциям на координатные оси. $$d_{Mh}(x,y)=\sum_{i=1}^N |x_i-y_i|$$
3. Минковского расстояние (Lp‑норма). Обобщение для Евклидова и Манхэттенского расстояния (p = 2 и 1 соответственно) $$d_{p}(x,y)= \left(\sum_{i=1}^N |x_i-y_i|^p\right)^{1 \over p}$$
4. Чебышёвское расстояние (L∞). Берется максимум от модуля разницы координат. Визуально проще всего представить в виде расстояний, которые может преодолет король на шахматной доске, поскольку для него перемещение по диагонали разноценно перемещению по оси координат. $$d_{∞}(x,y)=\max_{i=1,...,N} |x_i-y_i|$$
5. Хэммингово расстояние. Идейно схоже с Манхэттеновским растоянием, но применяется для двоичных векторов и отображает число раличающихся позиций
$$d_h(x,y)= count_{i=1}^N(x_i \ne y_i)$$
6. Косинусное расстояние (на основе косинусного сходства). Это мера, которая выводится из косинусного сходства (угла между векторами). В отличие от евклидова расстояния, она измеряет не длину, а разницу в направлении двух векторов. $$d_{\text{cos}}(x,y)=1-\frac{\sum_{i=1}^N x_i y_i}{\sqrt{\sum_{i=1}^N x_i^2}\sqrt{\sum_{i=1}^N y_i^2}}$$
7. Джаккардово расстояние. Это мера различия между множествами или бинарными векторами. Она показывает, насколько велика доля несовпадающих элементов относительно общего числа элементов.$$d_{\text{Jaccard}}(A,B)=1 - \frac{A \cap B}{A \cup B} $$

Для каждой меры укажите:
- тип данных (числовые / бинарные / категориальные / векторные), для которых она корректно применяется;
- особенности (чувствительность к масштабу, выбросам, разреженности).

1. Евклидово расстояние  
Тип данных: Числовые (количественные признаки, непрерывные).  
Особенности:  
    Чувствительно к масштабу (требует нормализации, если признаки имеют разные единицы измерения).  
    Сильно зависит от выбросов (квадрат разности увеличивает вклад больших отклонений).  
    Предполагает, что признаки некоррелированы и имеют одинаковую важность (если не используются веса).  


2. Манхэттенское расстояние  
    Тип данных: Числовые (количественные признаки).  
    Особенности:  
        Менее чувствительно к выбросам, чем евклидово (так как не возводит разности в квадрат).  
        Чувствительно к масштабу, также требует нормализации.  
        Хорошо работает в пространствах большой размерности  

3. Расстояние Минковского  
    Тип данных: Числовые.  
    Особенности:  
        Параметр p позволяет регулировать чувствительность к выбросам: чем больше p, тем больший вес получают большие отклонения  
        При p→∞ стремится к расстоянию Чебышёва.  
        Требует выбора оптимального p под задачу, но чаще всего используются p=1 или p=2.  

4. Расстояние Чебышёва  
    Тип данных: Числовые.  
    Особенности:  
        Игнорирует все координаты, кроме максимально различающейся.   
        Чувствительно к масштабу (по той единственной координате).  
        Полезно, когда важна максимальная разница (например, в задачах   планирования маршрутов, управлении роботами).  
        Устойчиво к выбросам по остальным координатам, но крайне   чувствительно к выбросу именно по максимуму.  


5. Хэммингово расстояние  
    Тип данных: Бинарные данные / бинарные векторы.  
    Особенности:  
        Важно для флагов, текстовых кодов  
        Все ошибки в равной степени влияют на ошибку  



6. Косинусное расстояние  
    Тип данных: Числовые векторы  
    Особенности:  
        Не чувствителен к длинам векторов (обращает внимание на угол   отклонения)  
        Частов применяется для сравнения текстов на схожесть.  

7. Джаккардово расстояние.  
    Тип данных: Множества признаков или тэгов  
    Особенности:  
        Принимает значения от 0 до 1  
        Не чувствительно к масштабам  
        Применяется для сравнения текстов

## Задача 1.2. Расстояния в бинарном пространстве и решающее правило

**Условие.**
Рассматривается пространство двух бинарных признаков $Z_1, Z_2 \in \{0,1\}.$

Дана обучающая выборка двух классов (образов) I и II:

| Z₁ | Z₂ | класс |
|----|----|-------|
| 0  | 0  | II    |
| 0  | 1  | II    |
| 1  | 0  | I     |
| 1  | 1  | I     |

1. Запишите **решающее правило** в словесной форме для разделения классов I и II.
2. Запишите это же правило в виде линейной дискриминантной функции $g(z) = w_0 + w_1 Z_1 + w_2 Z_2$ и порогового правила $g(z) \ge 0$ / $< 0$.
3. Используя хэммингово расстояние, сформулируйте правило ближайшего соседа с двумя прототипами:
   - прототип класса I: $z^{(I)} = (1,1)$,
   - прототип класса II: $z^{(II)} = (0,0)$.
   Объясните, к какому классу будут отнесены объекты с $(Z_1,Z_2) = (1,0)$ и $(0,1)$.


### Ваши ответы к задаче 1.2

_Заполните:_
1. Словесное решающее правило: Если $Z_1$ = 1, Класс I, иначе Класс II
2. Вид линейной функции g(z) и порог: $g(z) =  Z_1 $ $g(z) \ge 1 ~~—~~ \text{Класс I} / <1 ~~—~~ \text{Класс II} $
3. Правило ближайшего соседа по хэммингу и решения для (1,0), (0,1).
Прототип класса I — $z^{(I)}=(1,1)$
Прототип класса II — $z^{(II)}=(0,0)$
Используя Хэммингово расстояние пытаемся распределить объекты А и В с $(Z_1,Z_2) = (1,0)$ и $(0,1)$ по классам. Получаем:  
А: $d_{h_{z^{(I)}}} = 1$  $d_{h_{z^{(II)}}} = 1$  
B: $d_{h_{z^{(I)}}} = 1$  $d_{h_{z^{(II)}}} = 1$  
Распределить объекты А и В по классам используя исключительно хэммингово расстояние до прототипов не представляется возможным. 

```python
import numpy as np
import pandas as pd

def g(df: pd.DataFrame):
    return df['Z1'].map(lambda x: "I" if x >= 1 else "II")

# Все возможные векторы (Z1, Z2)
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

# Классы: 1 = I, 0 = II
y = np.array([0, 0, 1, 1])

df_train = pd.DataFrame(X, columns=["Z1", "Z2"])
df_train["class_label"] = y
df_train["class_name"] = df_train["class_label"].map({0: "II", 1: "I"})
df_train["classified"] = g(df_train)
df_train
```
